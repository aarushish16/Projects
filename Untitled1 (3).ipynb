```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Conv2D, MaxPooling2D, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.applications import ResNet50
```

**Reading Images and Labels**

```python
data = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/input3/images/styles.csv', on_bad_lines='skip')
image_folder = '/content/drive/MyDrive/ColabNotebooks/input3/images/images'
data = data.dropna()
data = pd.DataFrame(data)
data['articleType'] = data['articleType'].str.lower().str.replace(' ', '')
print(data.shape)
```

**Sampling data and balancing classes**

```python
article_types = data['articleType'].value_counts()[data['articleType'].value_counts() > 600].index.tolist()
print(f'{len(article_types)} classes are present in the range')

# Define categories and sample size
categories = article_types
n = 200  # Sample size

# Group data by 'articleType'
grouped_data = data.groupby('articleType')

# Sample and concatenate data for each category
samples = [grouped_data.get_group(cat).sample(n, random_state=42) for cat in categories]

# Concatenate all sampled DataFrames
all_samples = pd.concat(samples, ignore_index=True)
print(all_samples.shape)
```

**EDA**

```python
n = 10
def show_k_random(k, name):
    choices = np.random.choice(all_samples.index, k, replace=False)
    choices = all_samples.loc[choices, ['id', name]]
    fig, axs = plt.subplots(1, k, figsize=(15, 10))
    for this_name, id, ax in zip(choices[name], choices['id'], axs.ravel()):
        img_path = f"{image_folder}/{id}.jpg"
        img = plt.imread(img_path)
        ax.imshow(img, cmap='gray')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(f'{this_name}')
    plt.tight_layout()
    plt.show()

labels = ['articleType']
for col in labels:
    print(col)
    show_k_random(n, col)
print(f'Showing {n} classes of {len(article_types)}')
```

**Extracting Labels/Classes**

```python
all_samples['labels'] = all_samples['articleType']

le = LabelEncoder()
labels = le.fit_transform(all_samples['labels'])
print(labels.shape)
print(all_samples.shape)
```

**Train Test and Val split**

```python
#step1 : Split dataset into training and temporary (validation + test) sets
train_df, temp_df, train_labels, temp_labels = train_test_split(all_samples[['id', 'labels']], labels, test_size=0.2,random_state=42,stratify=labels)


# Step 2: Split the temporary set into validation and test sets
val_df, test_df, val_labels, test_labels = train_test_split(temp_df, temp_labels,test_size=0.5,random_state=42,stratify=temp_labels)
```

```python
# Image data generator
train_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1.0/255)
```

**Train Generator**

```python
#train gen
train_df['id'] = train_df['id'].astype(str).apply(lambda x: x if x.endswith('.jpg') else x + '.jpg')

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,         # DataFrame containing image paths and labels
    directory=image_folder,     # Directory where images are stored
    x_col='id',                 # Column in DataFrame with image filenames
    y_col='labels',             # Column in DataFrame with labels
    target_size=(128, 128),     # Resize images to this size
    batch_size=32,
    seed=42,                    # Number of images to return in each batch
    class_mode='categorical'    # Type of label encoding
)

# Verify that generator is working
for batch in train_generator:
    print(batch[0].shape)  # Print shape of images
    print(batch[1].shape)  # Print shape of labels
    break
```

**Val Generator**

```python
#value gen
val_df['id'] = val_df['id'].astype(str).apply(lambda x: x if x.endswith('.jpg') else x + '.jpg')

val_generator = val_datagen.flow_from_dataframe(
    val_df,
    directory=image_folder,
    x_col='id',
    y_col='labels',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical')

# Verify that generator is working
for batch in val_generator:
    print(batch[0].shape)  # Print shape of images
    print(batch[1].shape)  # Print shape of labels
    break

print(all_samples.shape)
```

**Test Generator**

```python
test_df['id'] = test_df['id'].astype(str).apply(lambda x: x if x.endswith('.jpg') else x + '.jpg')
test_generator = test_datagen.flow_from_dataframe(
    dataframe=test_df,
    directory=image_folder,
    x_col='id',
    y_col='labels',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    shuffle=False)  # No shuffling for evaluation/prediction)
```

**Building Model**

```python
input_shape = (128, 128, 3)
resnet_base = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
resnet_base.trainable = True
inputs = Input(shape=input_shape)
x = resnet_base(inputs, training=True)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)


outputs = Dense(len(le.classes_), activation='sigmoid')(x)
model = Model(inputs, outputs)
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
import tensorflow as tf

class Stop_Training(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        # Ensure 'val_accuracy' exists in logs and is not None before comparison
        if logs.get('val_accuracy') is not None and logs.get('val_accuracy') >= 0.90:
            print(f"\nReached 80% validation accuracy at epoch {epoch + 1}, stopping training!")
            self.model.stop_training = True  # Correct attribute for stopping training


# Create an instance of the callback
callbacks = Stop_Training()

# Train your model and include the callback
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=30,
    callbacks=[callbacks])
```

**Evaluating Model**

```python
val_loss, val_acc = model.evaluate(val_generator)
print(f'Validation Accuracy: {val_acc * 100:.2f}%')
print(f'Validation Loss: {val_loss * 100:.2f}%')
# model.save('my_resnet18_model.keras')
```

**Confusion Matrix and Test Accuracy**

```python
test_loss, test_accuracy = model.evaluate(test_generator)
print(f'Test Accuracy: {test_accuracy * 100:.2f}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix


y_true = test_generator.classes
y_pred = model.predict(test_generator)
y_pred_classes = y_pred.argmax(axis=1)


cm = confusion_matrix(y_true, y_pred_classes)
class_labels = test_generator.class_indices
index_to_class = {v: k for k, v in class_labels.items()}
class_names = [index_to_class[i] for i in range(len(index_to_class))]

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Test Data')
plt.show()
```
